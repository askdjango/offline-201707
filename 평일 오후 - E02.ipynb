## for

```python
for i in [1, 2, 3, 4, 5]:
    print(i)
```

```python
for name in {'Tom', 'Steve', 'Tom'}:
    print(name)
```

```python
for key in {'Tom': 10, 'Steve': 8, 'Tom': 9}:
    print(key)
```

```python
for value in {'Tom': 10, 'Steve': 8, 'Tom': 9}.values():
    print(value)
```

```python
for (key, value) in {'Tom': 10, 'Steve': 8, 'Tom': 9}.items():
    print(key, value)
```

```python
for i in range(1, 21):
    if i % 2 == 0:
        print('짝수 :', i)
    else:
        print('홀수 :', i)
    break
```

```python
for i in range(1, 21):
    if i % 2 == 0:
        print('짝수 :', i)
        break
    else:
        print('홀수 :', i)
```

```python
for i in range(2, 10):
    print('# {}단'.format(i))
    for j in range(1, 10):
        print('{} * {} = {}'.format(i, j, i*j))
    print()
```

## 함수

```python
number = 2

for i in range(1, 10):
    print('{} * {} = {}'.format(number, i, number*i))
    
number = 3

for i in range(1, 10):
    print('{} * {} = {}'.format(number, i, number*i))
    
number = 4

for i in range(1, 10):
    print(number, i, number*i)
```

```python
def gugudan(number):
    print('# {} 단'.format(number))
    for i in range(1, 10):
        print('{} * {} = {}'.format(number, i, number*i))
```

```python
gugudan(2)
gugudan(3)
gugudan(4)
```

## 인자

```python
'{0} {1} {2}'.format('a', 'b', 'c')
```

```python
'{2} {1} {2}'.format('a', 'b', 'c')
```

```python
'{var1} {var3} {var2}'.format(var1='a', var2='b', var3='c')
```

```python
'{} {var3} {var2}'.format('a', var2='b', var3='c')
```

```python
'{0} {var3} {var2}'.format('a', var2='b', var3='c')
```

```python
''.format('a', 'b', 'c', 'd', var5='e', var6='f')
```

```python
'{0} {3} {var5}'.format('a', 'b', 'c', 'd', var5='e', var6='f')
```

```python
args = ('a', 'b', 'c', 'd')
'{0} {3} {var5}'.format(*args, var5='e', var6='f')
```

```python
args = ('a', 'b', 'c', 'd')
kwargs = {'var5': 'e', 'var6': 'f'}
'{0} {3} {var5}'.format(*args, **kwargs)
```
